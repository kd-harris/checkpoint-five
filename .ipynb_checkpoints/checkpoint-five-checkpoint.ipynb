## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
import pandas as pd
import numpy as np
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

mushrooms_df = pd.read_csv('mo_mushrooms_cleaned_3.csv', index_col=None)
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
#pairplot

sb.pairplot(mushrooms_df)
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
# assign x and y variables
X = mushrooms_df['edibility']
y = mushrooms_df['month']


# create train and test sets
# divide the training and testing sets into a 7:3 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
X_train
```

```python
y_train
```

```python
import statsmodels.api as sm

# add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# fit the regression line using ordinary least square method
lr = sm.OLS(y_train, X_train_sm).fit()

# parameters
lr.params
```

```python
# list all the different parameters of the regression line 
lr.summary()
```

```python
# visualize the regression line
plt.scatter(X_train, y_train)
plt.plot(X_train, 8.0101 + -0.0556*X_train, 'r')
plt.show()
```

## Error Terms

Finally, plot your error terms!

```python
# Predicting y_value using training data of X
y_train_pred = lr.predict(X_train_sm)

# Creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)

# Plotting the histogram using the residual values
fig = plt.figure()
sb.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
plt.scatter(X_train,res)
plt.show()
```
